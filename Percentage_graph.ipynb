```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

DATA_DIR = '/content/drive/MyDrive/BarclaysReviews/data/'

final_merge_reviews_with_quarters = pd.read_csv(DATA_DIR+"final_merge_reviews_with_quarters.csv")

def plot_percentage_graph(df,column1,value,xlabel,ylabel,title,column2='quarter'):
  percentages = df[df[column1] == value].groupby(column2).size() / df.groupby(column2).size() * 100
      
  plt.plot(percentages.index, percentages.values) 
  plt.ylabel(ylabel)
  plt.xlabel(xlabel)
  plt.title(title)
  plt.xticks(rotation=45, ha='right')
  plt.show()

plot_percentage_graph(final_merge_reviews_with_quarters,'customer_sentiment','NEGATIVE','quarter','Percentage of negative Reviews','Percentage of negative Reviews by Quarter')

```

```python
plot_percentage_graph(final_merge_reviews_with_quarters,'customer_sentiment','POSITIVE','quarter','Percentage of positive Reviews','Percentage of positive Reviews by Quarter')
```
