```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

DATA_DIR = '/content/drive/MyDrive/BarclaysReviews/data/'

final_merge_reviews_with_quarters = pd.read_csv(DATA_DIR+"final_merge_reviews_with_quarters.csv")

def plot_percentage_graph(df,column1,value,xlabel,ylabel,title,column2='quarter',columns={}):
  print(columns)
  if bool(columns):
    # print("columns")
    for key in columns:
      # print(df.shape)
      df = df[df[key].str.contains(columns[key])]
  df['review_date'] = pd.to_datetime(df['review_date'])
  df = df[(df['review_date'] > "2022-01-31")]

  percentages = df[df[column1] == value].groupby(column2).size() / df.groupby(column2).size() * 100
      
  plt.plot(percentages.index, percentages.values) 
  plt.ylabel(ylabel)
  plt.xlabel(xlabel)
  plt.title(title)
  plt.xticks(rotation=45, ha='right')
  plt.ylim((0,100))
  plt.show()



```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


counts = final_merge_reviews_with_quarters.groupby(['quarter', 'rating']).size().reset_index(name='counts')

sns.lineplot(data=counts, x='quarter', y='counts', hue='rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```
